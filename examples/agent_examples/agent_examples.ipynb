````python
from IPython.display import display, HTML
display(HTML("<style>.container { width:80% !important; }</style>"))
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import levy_stable
from tqdm.notebook import tqdm
import numpy as np
from neuralplayground.arenas import Simple2D, ConnectedRooms
from neuralplayground.agents import AgentCore, RandomAgent, LevyFlightAgent, Weber2018
````

## Initialise agent

Initialise a random Levy Flight Agent

````python
agent = LevyFlightAgent()
````

## Initialise the arena

We first create a random 2D arena for the agent to interact with. For more information on how to create arena refer to the arena jupyter notebook.

````python
time_step_size = 0.1 #sec
agent_step_size = 3

# Init environment
env = Simple2D(time_step_size = time_step_size,
               agent_step_size = agent_step_size,
               arena_x_limits=(-100, 100), 
               arena_y_limits=(-100, 100))
````

## Agent and Arena Interaction

The agent receives observations from the environment (reward, visual cues, etc.) and uses these to take an action with the function ``` act```  which in turn will ``` update```  both its state and the state of the environment, generating new observations ```env.step``` . 

````python
n_steps = 10000

# Initialize environment
obs, state = env.reset()
for i in range(n_steps):
    # Observe to choose an action
    action = agent.act(obs)
    # Run environment for given action
    # print(action)
    obs, state, reward = env.step(action)
````

We plot the trajectory of the agent in the arena.

````python
ax = env.plot_trajectory()
ax.grid()
ax.set_xlabel("width", fontsize=18)
ax.set_ylabel("depth", fontsize=18)
# plt.savefig("two_rooms.pdf", bbox_inches="tight")
plt.show()
````
